```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")
```

```python
theta = np.arange(0, 6*np.pi + 0.1, 0.1)  # 0 to 6π with step 0.1
y = np.sin(theta / 3)  # y = sin(θ/3)
```

```python
theta.shape
```

```python
plt.figure(figsize=(10, 6))

# Plot the sine wave
plt.plot(theta, y, label=r'$y = \sin\left(\frac{\theta}{3}\right)$', color='blue')

# Add vertical lines to mark the period
plt.axvline(x=2*np.pi, color='red', linestyle='--', label=r'$2\pi$ (Standard Period for $\sin\theta$)')
plt.axvline(x=6*np.pi, color='green', linestyle='--', label=r'$6\pi$ (Actual Period for $\sin\left(\frac{\theta}{3}\right)$)')

# Customize axes to show multiples of π
plt.xticks(
    ticks=np.arange(0, 6.1*np.pi, np.pi),
    labels=['0', r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$', r'$5\pi$', r'$6\pi$']
)

# Labels and title
plt.xlabel(r'$\theta$ (radians)', fontsize=12)
plt.ylabel(r'$y = \sin\left(\frac{\theta}{3}\right)$', fontsize=12)
plt.title(r'Period of $y = \sin\left(\frac{\theta}{3}\right)$ is $6\pi$', fontsize=14)
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(3*x)

plt.figure(figsize=(10, 4))
plt.plot(x, y, label=r'$\sin(3x)$')
plt.scatter([n*np.pi/3 for n in range(7)], [0]*7, color='red', zorder=5, label='Solutions')
plt.axhline(0, color='black', linewidth=0.5)
plt.title(r'Solutions of $\sin(3x) = 0$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
