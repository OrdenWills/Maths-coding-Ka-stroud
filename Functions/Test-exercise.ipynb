```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.arange(0, 10, 0.2)
X
```

```python
y = 3*np.power(X, 4)
y
```

```python
y_inv = 3*np.power(X, 1/4)
y_inv
```

```python
df = pd.DataFrame({'X': X, 'y': y,})
df
```

```python
df_inv = pd.DataFrame({'X': X, 'y_inv': y_inv})
df_inv
```

```python
sns.scatterplot(data=df, x='X', y='y')
plt.grid()
plt.title("3*X^4")
plt.show()
```

```python
sns.scatterplot(data=df_inv, x='X', y='y_inv')
plt.grid()
plt.title("3*X^(1/4)")
plt.show()
```

```python
sns.scatterplot(data=df, x='X', y='y')
sns.scatterplot(data=df_inv, x='X', y='y_inv')
plt.ylim(-10, 100)
plt.grid()
plt.title("3*X^4 and 3*X^(1/4)")
plt.show()
```

```python
y1 = np.power(-X, 3)
y1
```

```python
df1 = pd.DataFrame({'X': X, 'y1': y1})
df1
```

```python
sns.scatterplot(data=df1, x='X', y='y1')
plt.grid()
plt.title("-X^3")
plt.show()
```

```python
y3 = np.power(-X, 1/3)
y3
```

```python
y4 = 3-np.power(X, 2)
y4
```

```python
df4 = pd.DataFrame({'X': X, 'y4': y4})
df4
```

```python
sns.scatterplot(data=df4, x='X', y='y4')
plt.grid()
plt.title("3-X^2")
plt.show()
```

```python
y5 = 3-np.power(X, 1/2)
y5
```

```python
df5 = pd.DataFrame({'X': X, 'y5': y5})
df5
```

```python
sns.scatterplot(data=df5, x='X', y='y5')
plt.grid()
plt.title("3-X^(1/2)")
plt.show()
```

```python
sns.scatterplot(data=df4, x='X', y='y4')
sns.scatterplot(data=df5, x='X', y='y5')
# plt.plot(X,X)
plt.ylim(-10, 10)
plt.grid()
plt.title("-X^3 and 3-X^(1/2)")
plt.show()
```
