```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(-1,5)
y = np.array(np.power(X,3)- (6*np.power(X,2)) + np.multiply(9,X) + 6)

data = pd.DataFrame({'x':X,"y":y})
data
```

```python
plt.figure(figsize=(7,5))
plt.plot(X,y,'*')
plt.plot(X,y,'--')
plt.scatter(1, [10]*1, color='red', zorder=5, label='maximum')
plt.scatter([3]*1, [6]*1, color='green', zorder=5, label='minimum')
plt.axvline(1, color='blue',linestyle='--', linewidth=0.8)
plt.axvline(3, color='blue',linestyle='--', linewidth=0.8)
plt.axhline(6,color='blue',linestyle='--',linewidth=0.8)
plt.axhline(8,color='blue',linestyle='--',linewidth=0.5)
plt.scatter(2,8,color='yellow',zorder=3,label='inflex')
plt.xlabel('xlabel')
plt.ylabel("ylable")
plt.legend()
plt.grid(True)
plt.title("plot of y=X^3-6X^2+9x+6")
plt.show()
```
