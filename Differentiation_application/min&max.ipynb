```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
x = np.arange(-3,5)

y = np.array((np.power(x,3)/3)-(np.power(x,2)/2)- np.multiply(2,x)+5)

# ymin = np.array((np.power(2,3)/3)-(np.power(2,2)/2)- np.multiply(2,2)+5)
# ymax = np.array((np.power(-1,3)/3)-(np.power(-1,2)/2)- np.multiply(-1,2)+5)

print("ymax = {} and ymin = {}".format(y.max(),y.min()))
```

```python
data = pd.DataFrame({"x":x,"y":y})
data
```

```python

plt.figure(figsize=(7,4))
plt.plot(x,y,'*',color='red')
plt.plot(x,y,'--')
plt.xlabel('x')
plt.ylabel("y")
plt.title('Graph of y=f(x) showing max at x=-1 and min at x=2')
plt.grid()
plt.legend()
plt.show()
```
