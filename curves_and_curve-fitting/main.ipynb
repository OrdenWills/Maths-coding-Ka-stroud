```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### LOG Curves

```python
x = np.arange(1,1000)
y = np.log(x)


data = pd.DataFrame({'x':x,'y':y})

data.head(15)
```

```python
data.tail(15)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y,zorder=3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y=logx')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y,zorder=3,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0,100])
plt.title('Graph of y=logx')
plt.grid()
plt.show()
```

### Exp Curve

```python
X = np.arange(0,14)
y = np.exp(X)


data = pd.DataFrame({'X':X,"y":y})
data
```

```python
plt.figure(figsize=(8,4))
plt.scatter(X,y,zorder=3, color='green')
plt.xlabel('X')
plt.ylabel('y')
plt.title('y=exp^x')
plt.grid()
plt.show()
```

```python
y = np.exp(-X)

data2 = pd.DataFrame({'X':X,"y":y})
data2
```

```python
plt.figure(figsize=(8,4))
plt.scatter(X,y,zorder=3, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.title('y=exp^-x')
plt.show()
```
