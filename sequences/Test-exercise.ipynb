```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,30)
y = np.array(3-np.multiply(2,X))


data = pd.DataFrame({"x":X,"y":y})
data.head()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x=data['x'],y=data['y'],zorder=3,color='red')
plt.grid()
plt.axhline(0,color='black',linewidth=0.9)
plt.title("f(x)=3-2x x>0")
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
y2 = np.multiply(2,-1^X)
data2 = pd.DataFrame({"x":X,"y":y2})
data2.head()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x=data2['x'],y=data2['y'],zorder=3,color='red')
plt.grid()
plt.axhline(0,color='black',linewidth=0.9)
plt.title("f(x)=2(-1)^X")
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
X1 = np.arange(1,55)
y3 = np.divide(-1^X1,X1)

data3 = pd.DataFrame({'x':X1,"y":y3})

data3.tail(17)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x=data3['x'],y=data3['y'],color='red',zorder=3)
plt.grid()
plt.axhline(0,linewidth=0.8,color='black')
plt.xlabel("xlabel")
plt.ylabel('ylabel')
plt.title('f(x)=(-1)^x/x  ')
plt.show()
```
