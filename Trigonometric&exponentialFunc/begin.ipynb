```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.arange(-10, 10, 1)
y = np.sin(X)

df = pd.DataFrame({'X': X, 'y': y})
df
```

```python
plt.plot(df['X'], df['y'])
plt.show()
```

```python
sns.scatterplot(x='X', y='y', data=df)
plt.grid()
plt.show()
```

```python
y1 = np.cos(X)
df1 = pd.DataFrame({'X': X, 'y': y1})
sns.scatterplot(x='X', y='y', data=df1)
plt.grid()
plt.show()
```

```python
df1
```

```python
plt.plot(df['X'], df['y'],label='sin x')
plt.plot(df1['X'], df1['y'],label='cos x')
plt.legend()
plt.grid()
plt.show()
```
