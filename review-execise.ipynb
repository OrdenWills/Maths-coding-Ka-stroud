```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.round(np.arange(-2,6,0.4),decimals=2)
X
```

```python
y = np.round(np.array((X**2-3*X + 2)),decimals=2)
y
```

```python
df = pd.DataFrame({'x':X,'y':y})
df.head(11)
```

```python
y.max()
```

```python
np.where(y==0)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X,y,'*')
plt.plot(X,y,'-')
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')
# plt.ylim([y.min(),y.max()],)
plt.title('y = x**2-3x+2')
plt.show()
```

```python
y1 = np.round(np.array(X**2-1),decimals=2)
y1
```

```python
df1 = pd.DataFrame({'x':X,"y":y1})
df1.head(12)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X,y1,'*')
plt.plot(X,y1)
plt.title("X**2-1")
plt.xlabel('x')
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
y2 = np.round(np.array((4*X**2-3*X+25)),decimals=2)
y2
```

```python
df2 = pd.DataFrame({'x':X,'y':y2})
df2
```

```python
plt.figure(figsize=(8,6))
plt.plot(X,y2,'*')
plt.plot(X,y2,'-')
plt.grid()
plt.title('4*X**2-3*X+25')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y3 = np.round(np.array((-X**3+6*X**2-8*X)),decimals=2)
y3
```

```python
df3 = pd.DataFrame({'x':X,'y':y3})
df3
```

```python
plt.figure(figsize=(8,6))
plt.plot(X,y3,'*')
plt.plot(X,y3,'-')
plt.title("-X**3+6*X**2-8*X")
plt.grid()
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
